```python
from icrawler.builtin import GoogleImageCrawler
import os

# Define class names and refined search queries
classes = {
    "drought_land": "drought land",
    "irrigated_land": "irrigated land",
    "lowland_irrigated_agricultural_land": "lowland irrigated agricultural land",
    "intensively_irrigated_plains_land": "intensively irrigated plains land",
    "upland_rainfed_agricultural_land": "upland rainfed agricultural land"
}

# Base folder
base_folder = "land_images_icrawler"

if not os.path.exists(base_folder):
    os.makedirs(base_folder)

# Function to download images for each class
def download_images(class_name, query):
    class_folder = os.path.join(base_folder, class_name)
    if not os.path.exists(class_folder):
        os.makedirs(class_folder)

    crawler = GoogleImageCrawler(storage={"root_dir": class_folder})
    crawler.crawl(keyword=query, max_num=300, filters={"type": "photo"})

# Loop through and download
for class_name, query in classes.items():
    print(f"Downloading images for: {class_name}")
    download_images(class_name, query)

print("All images downloaded successfully.")
```

```python
import os
import shutil
from google.colab import files

# --- Zip the whole folder ---
zip_filename = "land_images_icrawler.zip"
shutil.make_archive("land_images_icrawler", 'zip', base_folder)
print(f"Zipped all images into {zip_filename}")

# --- Download zip file to local device ---
files.download(zip_filename)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Parameters
img_size = (128, 128)  # Resize all images to 128x128
batch_size = 32

# Data Augmentation and Image Loading
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2  # 80% train, 20% validation
)

train_generator = datagen.flow_from_directory(
    base_folder,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

val_generator = datagen.flow_from_directory(
    base_folder,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

# Save class indices for later use
class_indices = train_generator.class_indices
print("Class Indices:", class_indices)

# Save for inference later
import json
with open('class_indices.json', 'w') as f:
    json.dump(class_indices, f)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Define simple CNN
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(classes), activation='softmax')  # 5 classes
])

# Compile
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train
history = model.fit(train_generator, validation_data=val_generator, epochs=20)

# Save model
model.save('land_classification_model.h5')
print("Model saved.")
```

```python
from google.colab import files
from PIL import Image
import numpy as np

# Load trained model and class indices
model = tf.keras.models.load_model('land_classification_model.h5')
with open('class_indices.json', 'r') as f:
    class_indices = json.load(f)

# Reverse class indices for prediction
class_names = {v: k for k, v in class_indices.items()}

# Upload image from user
uploaded = files.upload()
img_path = list(uploaded.keys())[0]

# Preprocess user-uploaded image
def preprocess_image(img_path):
    img = Image.open(img_path).resize(img_size)
    img = np.array(img) / 255.0
    img = np.expand_dims(img, axis=0)
    return img

img = preprocess_image(img_path)

# Predict class
prediction = model.predict(img)
predicted_class = np.argmax(prediction)
predicted_label = class_names[predicted_class]

print(f"Predicted Land Type: {predicted_label}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to display images with predictions (green for correct, red for wrong)
def plot_predictions(generator, model, class_names, num_images=20):
    images, labels = next(generator)  # Grab one batch from the validation set
    predictions = model.predict(images)

    plt.figure(figsize=(15, 15))

    for i in range(num_images):
        ax = plt.subplot(5, 5, i + 1)

        img = images[i]
        true_label = np.argmax(labels[i])
        predicted_label = np.argmax(predictions[i])

        # Get class names
        true_class = class_names[true_label]
        predicted_class = class_names[predicted_label]

        # Show image
        plt.imshow(img)
        plt.axis("off")

        # Display prediction result with color
        color = "green" if predicted_label == true_label else "red"
        plt.title(f"Pred: {predicted_class}\nTrue: {true_class}", color=color)

    plt.tight_layout()
    plt.show()

# Reverse class indices (you already saved them in class_indices.json earlier)
class_names = {v: k for k, v in class_indices.items()}

# Plot predictions
plot_predictions(val_generator, model, class_names)
```
